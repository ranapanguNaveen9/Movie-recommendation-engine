```python
import pandas as pd
import numpy as np
```

```python
# 1930
df1930s = pd.read_html("https://en.wikipedia.org/wiki/List_of_Telugu_films_of_the_1930s")
df1930s = df1930s[2]
```

```python
df1930s.shape
```

```python
df1930s.columns
```

```python
df1930s = df1930s[['Title', 'Director', 'Cast', 'Genre']]
```

```python
df1930s["Music"] = np.NaN
```

```python
df1930s['year'] = np.NaN
```

```python
for i in range(0,len(df1930s)):
    if df1930s['Title'][i].isnumeric():
        a = df1930s['Title'][i]
    df1930s['year'][i] = a
```

```python
df1930s
```

```python
# removing numerics from title columns
df1930s = df1930s[pd.to_numeric(df1930s['Title'], errors='coerce').isna()]
df1930s = df1930s.reset_index(drop = True)
```

```python
one_list = []
for i in range(0,60):
    df = pd.read_html("https://en.wikipedia.org/wiki/List_of_Telugu_films_of_"+str(1940+i),
                      match = "Title")
    df = df[0]
    #if "Music" | "Music director" | "Music Director" not in df.columns:
     #   df["Music"] = np.NaN
    if "Genre" not in df.columns:
        df["Genre"] = np.NaN
    if "Music Director" in df.columns:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music Director']]
            df.columns = ['Title', 'Director', 'Cast','Genre', 'Music']
            df['year'] = 1940+i
    elif "Music" not in df.columns:
        df['Music'] = np.NaN
        df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
        df['year'] = 1940+i 
    else:
        try:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music director']]
            df.columns = ['Title', 'Director', 'Cast', 'Music']
            df['year'] = 1940+i
        except:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
            df['year'] = 1940+i
    one_list.append(df)
```

```python
df4099 = pd.concat(one_list,ignore_index = True, sort = False)
```

```python
df4099
```

```python
two_list = []
for i in range(0,20):
    df = pd.read_html("https://en.wikipedia.org/wiki/List_of_Telugu_films_of_"+str(2000+i),
                      match = "Title")
    #df = df[0]
    k = []
    for j in range(0,len(df)-1):
        if "Rank" not in df[j].columns:
            k.append(df[j])
    df = pd.concat(k,ignore_index = True,sort = False)
    if "Genre" not in df.columns:
        df["Genre"] = np.NaN
    if "Music Director" in df.columns:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music Director']]
            df.columns = ['Title', 'Director', 'Cast','Genre', 'Music']
            df['year'] = 2000+i
    elif "Music" not in df.columns:
        df['Music'] = np.NaN
        df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
        df['year'] = 2000+i 
    else:
        try:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music director']]
            df.columns = ['Title', 'Director', 'Cast', 'Music']
            df['year'] = 2000+i
        except:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
            df['year'] = 2000+i
    two_list.append(df)
```

```python
df0019 = pd.concat(two_list,ignore_index = True, sort = False)
```

```python
df0019
```

```python
three_list = []
for i in range(0,2):
    df = pd.read_html("https://en.wikipedia.org/wiki/List_of_Telugu_films_of_"+str(2020+i),
                      match = "Title")
    k = []
    for j in range(0,len(df)):
        if "Rank" not in df[j].columns:
            k.append(df[j])
    df = pd.concat(k,ignore_index = True,sort = False)
    if "Music" not in df.columns:
        df['Music'] = np.NaN
    if "Genre" not in df.columns:
        df["Genre"] = np.NaN
        
    df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
    df['year'] = 2020+i
    three_list.append(df)
```

```python
df2021 = pd.concat(three_list,ignore_index = True, sort = False)
```

```python
df2021
```

```python
final_list = [df1930s,df4099,df0019,df2021]
```

```python
tollywood = pd.concat(final_list,ignore_index = True, sort = False)
```

```python
tollywood = tollywood[tollywood['Title'].notna()]
tollywood = tollywood.reset_index(drop = True)
```

```python
tollywood.info()
```

```python
tollywood['language'] = "Telugu"
tollywood['industry'] = "Tollywood"
```

```python
tollywood = tollywood.drop_duplicates()
tollywood = tollywood.reset_index(drop = True)
```

```python
tollywood.to_csv("Tollywood_movies.csv",header =  True,index = False)
```
