```python
import pandas as pd
import numpy as np
```

```python
# 1930
df1930s = pd.read_html("https://en.wikipedia.org/wiki/List_of_Tamil_films_of_the_1930s")
```

```python
li = []

for i in range(0,9):
    df1930s[1+i]['year'] = 1931+i
    li.append(df1930s[1+i])
```

```python
df1930s = pd.concat(li,ignore_index = True)
```

```python
df1930s["Genre"] = np.NaN
```

```python
df1930s = df1930s[['Title', 'Director', 'Cast', "Genre",'Music','year']]
```

```python
# 2004
df2004 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Tamil_films_of_2004",
                      match = "Opening")

df2004 = df2004[0]

df2004.columns = df2004.iloc[0]

# drop first row
df2004.drop(index=df2004.index[0], 
        axis=0, 
        inplace=True)

df2004 = df2004.reset_index(drop =  True)

df2004['year'] = 2004
```

```python
df2004.columns
```

```python
df2004['Genre'] = np.NaN
```

```python
df2004 = df2004[['Title', 'Director', 'Cast','Genre','Music Director', 'year']]
```

```python
df2004.columns = ['Title', 'Director', 'Cast','Genre', 'Music', 'year']
```

```python
df2004
```

```python
df2011 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Tamil_films_of_2011",
                      match = "Opening")

a = df2011[0]
a = a[['Title', 'Director',"Actors",'Genre']]
a.columns = ['Title', 'Director',"Cast", 'Genre']

b = df2011[1]
b = b[['Title', 'Director',"Cast",'Genre']]
```

```python
df2011 = pd.concat([a,b],ignore_index = True,sort=False)
df2011['Music'] = np.NaN
df2011['year'] = 2011
```

```python
df2011.columns
```

```python
df2011
```

```python
final_list = [df1930s,df4021,df2004,df2011]
```

```python
tamil_movies = pd.concat(final_list,ignore_index = True, sort = False)
```

```python
tamil_movies = tamil_movies[tamil_movies['Title'].notna()]
tamil_movies = tamil_movies.reset_index(drop = True)
```

```python
tamil_movies['language'] = "Tamil"
tamil_movies['industry'] = "Kollywood"
```

```python
tamil_movies
```

```python
tamil_movies = tamil_movies.drop_duplicates()
tamil_movies = tamil_movies.reset_index(drop = True)
```

```python
tamil_movies.to_csv("Kollywood_movies.csv",header =  True,index = False)
```
