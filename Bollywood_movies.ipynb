```python
import pandas as pd
import numpy as np
```

```python
one_list = []
for i in range(0,102):
    df = pd.read_html("https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_"+str(1920+i),
                      match = "Director")
    
    if len(df)>1:
        k = []
        for j in range(0,len(df)):
            if "Original film" not in df[j].columns:
                if "No." not in df[j].columns:
                    if "Worldwide Gross" not in df[j].columns:
                        k.append(df[j])
        df = pd.concat(k,ignore_index = True,sort = False)
        if "Genre" not in df.columns:
            if "Music" not in df.columns:
                df['Music'] = np.NaN
                df['Genre'] = np.NaN
                df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
                df['year'] = 1920+i
                print(1920+i)
                print(df.columns)
        elif "Music Director" in df.columns:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music Director']]
            df.columns = ['Title', 'Director', 'Cast','Genre', 'Music']
            df['year'] = 1920+i
            print(1920+i)
            print(df.columns)
        elif "Music" not in df.columns:
            df['Music'] = np.NaN
            df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
            df['year'] = 1920+i
            print(1920+i)
            print(df.columns)
        elif "Genre" not in df.columns:
            df['Genre'] = np.NaN
            df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
            df['year'] = 1920+i
            print(1920+i)
            print(df.columns)
        else:
            try:
                df = df[['Title', 'Director', 'Cast','Genre', 'Music director']]
                df.columns = ['Title', 'Director', 'Cast','Genre', 'Music']
                df['year'] = 1920+i
                print(1920+i)
                print(df.columns)
            except:
                df = df[['Title', 'Director', 'Cast','Genre','Music']]
                df['year'] = 1920+i
                print(1920+i)
                print(df.columns)
    else:
        df = df[0]
        if "Music Director" in df.columns:
            df = df[['Title', 'Director', 'Cast','Genre', 'Music Director']]
            df.columns = ['Title', 'Director', 'Cast','Genre', 'Music']
            df['year'] = 1920+i
            print(1920+i)
            print(df.columns)
        elif "Music" not in df.columns:
            df['Music'] = np.NaN
            df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
            df['year'] = 1920+i
            print(1920+i)
            print(df.columns)
        elif "Genre" not in df.columns:
            df['Genre'] = np.NaN
            df = df[['Title', 'Director', 'Cast','Genre', 'Music']]
            df['year'] = 1920+i
            print(1920+i)
            print(df.columns)
        else:
            try:
                df = df[['Title', 'Director', 'Cast','Genre', 'Music director']]
                df.columns = ['Title', 'Director', 'Cast','Genre', 'Music']
                df['year'] = 1920+i
                print(1920+i)
                print(df.columns)
            except:
                df = df[['Title', 'Director', 'Cast','Genre','Music']]
                df['year'] = 1920+i
                print(1920+i)
                print(df.columns)
    one_list.append(df)
```

```python
bollywood = pd.concat(one_list,ignore_index = True,sort = False)
```

```python
bollywood
```

```python
bollywood = bollywood[bollywood['Title'].notna()]
bollywood = bollywood.reset_index(drop = True)
```

```python
bollywood['language'] = "Hindi"
bollywood['industry'] = "Bollywood"
```

```python
bollywood = bollywood.drop_duplicates()
bollywood = bollywood.reset_index(drop = True)
```

```python
bollywood.to_csv("Bollywood_movies.csv",header =  True,index = False)
```
